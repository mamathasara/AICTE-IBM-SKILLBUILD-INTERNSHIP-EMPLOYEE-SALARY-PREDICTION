```python
 #Replacing the name of the category with some other suitable name
 data['occupation'] = data['occupation'].replace({'?': 'Others'})
```

```python
print(data['occupation'].value_counts())
```

```python
data
```

```python
data['workclass'] = data['workclass'].replace({'?': 'Not Listed'})
print(data['workclass'].value_counts())
```

```python
#Removing unecessary categories from the column
data = data[data['workclass']!='Without-pay']
data = data[data['workclass']!='Never-worked']
```

```python
print(data['workclass'].value_counts())
```

```python
data.shape
```

```python
data = data[data['education']!='5th-6th']
data = data[data['education']!='1st-4th']
data = data[data['education']!='Preschool']
print(data['education'].value_counts())
```

```python
#Redundancy (Removing Unnecessary columns)
data = data.drop('education', axis=1)
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
# List of columns to encode
cols_to_encode = [
'workclass', 'marital-status', 'occupation',
'relationship', 'race', 'gender', 'native-country'
]
# Encode each column using a fresh LabelEncoder
for col in cols_to_encode:
    le = LabelEncoder()
    data.loc[:,col] = le.fit_transform(data[col])
```
